```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("abundance.csv/abundance.csv")
```

```python
renewed_data = df.loc[(df["disease"]!='nd') & (df["disease"]!='n_relative') & (df["disease"]!='y') & (df["disease"]!='-') & (df["disease"]!=' -')]
renewed_data2 = renewed_data.assign(d_factor = (renewed_data["disease"] == 'n'))
stool_data = renewed_data2.loc[(renewed_data2["bodysite"]=='stool')]
```

```python
stool_data
```

```python
cstool_data = stool_data.iloc[:,211:3513]
cstool_data = cstool_data.loc[:,((cstool_data)!=0).sum()>10]
```

```python
cstool_data = cstool_data.loc[:,cstool_data.columns.str.contains("s__")]
```

```python
cstool_data
```

```python
data = cstool_data
corr = data.corr()
```

```python
corr.shape
```

```python
corr
```

```python
fig = plt.figure(figsize=(50, 50))
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(data.columns),1)

plt.show()
```

```python
import scipy
import scipy.cluster.hierarchy as sch

X = corr.values
d = sch.distance.pdist(X)   
L = sch.linkage(d, method='complete')
ind = sch.fcluster(L, 0.5*d.max(), 'distance')
columns = [cstool_data.columns.tolist()[i] for i in list((np.argsort(ind)))]


```

```python
data = data.reindex(columns, axis=1)
corr2 = data.corr()
```

```python
data.shape
```

```python
fig = plt.figure(figsize=(100, 100))
ax = fig.add_subplot(111)
cax = ax.matshow(corr2,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(data.columns),1)

plt.show()
```

```python
X = corr.values
d = sch.distance.pdist(X)   
L = sch.linkage(d, method='complete')
ind = sch.fcluster(L, 0.55*d.max(), 'distance')
columns = [cstool_data.columns.tolist()[i] for i in list((np.argsort(ind)))]
```

```python
np.unique(ind)
```

```python
data = data.reindex(columns, axis=1)
```

```python
corr4 = data.corr()
```

```python
ind2 = np.sort(ind)
corr4["cluster"] = ind2.tolist()
for i in np.unique(ind):
    corr4["cluster" + str(i)] = (corr4["cluster"] ==i).astype(float)
    corr4["acluster" + str(i)] = (corr4["cluster"] ==i).astype(float)
    corr4["bcluster" + str(i)] = (corr4["cluster"] ==i).astype(float)
```
